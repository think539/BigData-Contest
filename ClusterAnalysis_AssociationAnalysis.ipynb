### MLxtend의 Apriori와 Association_rules를 이용하여 연관분석

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
frequent_itemsets = apriori(data1, use_colnames=True,min_support=0.001,)
```

```python
frequent_itemsets.sort_values(by='support',ascending=False)
```

```python
rules=association_rules(frequent_itemsets,metric='lift')
```

```python
df=rules[rules.consequents.apply(lambda x: len(x)==1)]
```

```python
df
```

```python
df=df.sort_values(by='confidence',ascending=False)
```

DSRCalc(DSR계산기)의 경우 support와 confidence가 낮아 분석에서 제외  
  
일반적으로 Apriori 알고리즘의 경우 기본 support가 0.2 이상인 경우만 사용하지만, DSR계산기, 여윳돈계산기의 지지도가 낮아 표시되지 않았고,  
상기한 이유로 min_support=0.001를 지정
